```python
import turtle
import time


def koch(t, len, depth):

    len = len // 3

    if depth > 1:
        depth -= 1

        koch(t, len, depth)
        t.left(60)
        koch(t, len, depth)
        t.right(120)
        koch(t, len, depth)
        t.left(60)
        koch(t, len, depth)

    else:

        t.forward(len)
        t.left(60)
        t.forward(len)
        t.right(120)
        t.forward(len)
        t.left(60)
        t.forward(len)


screen = turtle.Screen()
screen.setup(1000, 500, 0, -50)
t = turtle.Turtle()
t.hideturtle()
t.color('red')
turtle.tracer(1, 0)


len = 500
depth = 2

koch(t, len, depth)

turtle.done()

#строим таблицу
for i in range (1, 5):
    turtle.TurtleScreen._RUNNING = True
    turtle_time = turtle.Turtle()
    turtle_time.hideturtle()
    turtle.tracer(2, 0)

    start = time.time()
    koch(turtle_time, len, i)
    finish = time.time() - start
    print("Текущая глубина: {} - Time: {}".format(i, "%8.8f"%finish))
```
